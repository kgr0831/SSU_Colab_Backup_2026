```python
hp = 200
defence = 15
monDmg = 30
hitCount = 3

dmg = monDmg - defence
totalDmg = dmg * hitCount

print(f"남은 체력 : {hp - totalDmg}")
```

```python
# 기본기 실습 : 복리계산
myMoney = 5_000_000
rate = .05

for _ in range(5) :
  myMoney += (myMoney * rate) # myMoney = myMoney + (myMoney * rate)

print(f"5년후 총 수령액 : {round(myMoney):,}원")
```

```python
# 변수 설정

payment = 10000
price = 3750
change = payment - price

coinList = [500, 100, 50, 10]
results = []

# 동전 개수 계산
for coin in coinList :
    result = change // coin
    print(coin)
    change %= coin
    results.append(result)

# 결과 출력
for i in range(len(coinList))  :
    print(f"{coinList[i]}원 동전 : {results[i]}개")
```

```python
# 기본기 실습 : input() => 정수변환

sales = []

for i in range(3) :
    sales.append(int(input("{i + 1}월 매출")))

total = 0

for i in range(len(sales)) :
    print(f"{i + 1}월 매출 : {sales[i]}")
    total += sales[i]

    print(f"총 매출 : {total}")
```

```python
weight = float(input("몸무게(kg)를 입력하세요 : "))
height = float(input("신장(m)를 입력하세요 : ")) / 100

bmi = weight / height ** 2
print(f"몸무게 : {weight}kg\n신장 : {height}m\nBMI : {"%.2f" %bmi}")
```

```python
# 직접해보기 : 영수증 출력기

name = input("상품명 : ")
price = int(input("단가(원) : "))
cnt = int(input("수량 : "))

total = price * cnt
vat = int(total * .1)
totalWVat = total + vat

print("===========================")

print(f"상품명 : {name}")
print(f"단가 : {price}원")
print(f"수량 : {cnt}개")
print(f"소계 : {total}원")
print(f"부가세(10%) : {vat}원")
print(f"최종 금액 : {totalWVat}원")
```
