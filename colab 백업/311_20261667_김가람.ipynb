#### 문제 정의 : 학생이 하루에 영어 공부를 20분 하는데, 7일동안 공부한다면 총 몇 분 공부할까요???

1) 문제 분해 : 영어공부를 / 하루 20분함 / 7일동안 함
2) 패턴 인식 : 하루 20분 공부 / 7일간
3) 추상화 : 20분 / 7시간
4) 알고리즘 : 공부 시간 변수 time = 20 / 공부 일 변수 date = 7 / 총 공부일 변수 total = time * date -> print(total)

```python
bread = 97
cnt = 3
maxCnt = bread // cnt
print(f"빵을 나누어 줄 수 있는 수 : {maxCnt}")
rest = bread % cnt
print(f"남은 빵의 수 : {rest}")
```
